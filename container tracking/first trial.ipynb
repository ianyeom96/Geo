```python
df_l = df_m[['latitude', 'longitude']]
df_l
```

```python
df_l = df_m[['latitude', 'longitude']]
df_ll = df_l.iloc[1:10]
df_ll
```

```python
latitude = 40.210033

longitude = 16.363449
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[latitude, longitude],
    zoom_start=3.5
)
coords = df_ll[['latitude', 'longitude']]


marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['latitude'], coords['longitude']):
    folium.Marker(
    location = [51.885, 4.2867],
    popup = 'Vessel departed from Rotterdam',
    icon = folium.Icon(color='red', icon='star')
).add_to(m)
    folium.Marker(
    location = [21.4961, 38.8076],
    popup = 'Vessel arrived in Jeddah',
    icon = folium.Icon(color='red', icon='star')
).add_to(m)
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)

m
```
