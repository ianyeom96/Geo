```python
sub_df = df[['사업장명','영업상태명', 'WGS84위도', 'WGS84경도']] #대괄호 두개 => 다중 열
sub_df = sub_df.dropna()
sub_df
```

```python
sub_df['영업상태명'].isin(['영업', '영업 등'])      #isin() 원소의 유무 판단
```

```python
sub_df[sub_df['영업상태명'].isin(['영업', '운영중'])]
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[latitude, longitude],
    zoom_start=10
)

coords = sub_df[['WGS84위도', 'WGS84경도']]


marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['WGS84위도'], coords['WGS84경도']):                     #all the latitudes into lat and all the longitudes into long to make markers on them
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```

```python
folium.CircleMarker([latitude, longitude],
                    color='blue',
                    radius = 120, 
                    tooltip='편의점 밀집지역').add_to(m)
m
```
